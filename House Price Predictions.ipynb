```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.svm import SVC
from xgboost import XGBRegressor
```

# Import the data set.

1. The training data columns are having value as 'NA' which is valid and we need to treat that.
2. in order to treat that we will be performing following operations

```python
df_train = pd.read_csv('train.csv')
df_test  = pd.read_csv('test.csv')
df_train
```

```python
df_train.shape,df_test.shape
```

# Outliers Detection and removing
- (1-2-3) model used for this and kept the z-Scoring between -3 and 3 

```python
z_Score = (df_train['SalePrice'] - df_train['SalePrice'].mean())/df_train['SalePrice'].std()
    
df_train = df_train[((z_Score >-3) & (z_Score <=3))]
df_train
```

```python
ls = ['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','Fence','MiscFeature','PoolQC']
nonNaCol={}
for i in df_train.columns.drop('SalePrice'):
    if i not in ls:
        nonNaCol[i] = ['', '#N/A', '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN', '-NaN', '-nan','1.#IND', '1.#QNAN', '<NA>', 'N/A', 'NA', 'NULL', 'NaN', 'n/a','nan', 'null']
    else:
        nonNaCol[i] = ['', '#N/A', '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN', '-NaN', '-nan','1.#IND', '1.#QNAN', '<NA>', 'N/A', 'NULL', 'NaN', 'n/a','nan', 'null']

df_train = pd.read_csv('train.csv',na_values=nonNaCol,keep_default_na=False)
```

# Find out the Numeric and Categorical columns 

1. To find this we will be carrying out below steps
    Use select_dtypes 
2. With impurities find out the Numerical and categorical data     

```python
columns_cat = list(df_train.select_dtypes(include='object').columns)
columns_num = list(df_train.select_dtypes(exclude='object').columns)

columns_num.remove('Id')
columns_num.remove('SalePrice')
```

# Fill up the missing data.
1. Using SimpleImputer fill up the missing data for numeric and categorical columns
2. for numeric we will be using median
3. for categorical columns we will be using mode(most_frequent)

```python
imputer_num = SimpleImputer(strategy='median')
imputer_cat = SimpleImputer(strategy='most_frequent')

imputer_num.fit(df_train[columns_num])
imputer_cat.fit(df_train[columns_cat])

df_train[columns_num] = imputer_num.transform(df_train[columns_num])
df_train[columns_cat] = imputer_cat.transform(df_train[columns_cat])

df_test[columns_num] = imputer_num.transform(df_test[columns_num])
df_test[columns_cat] = imputer_cat.transform(df_test[columns_cat])

print('**************************************************************************************************')
print('Checking if both train and test data are filled up correctly')

print('Training', df_train.isna().sum()[df_train.isna().sum()>0])
print('Testing', df_test.isna().sum()[df_test.isna().sum()>0])

print('**************************************************************************************************')
```

# Scaling the Numerical data 
1. Using MinMaxScaler we will be scaling the data

```python
scaler = MinMaxScaler()
scaler.fit(df_train[columns_num])

df_train[columns_num] = scaler.transform(df_train[columns_num])
df_test[columns_num] = scaler.transform(df_test[columns_num])
```

# Feature Enconding using OneHotEncoder

```python
ohe = OneHotEncoder(handle_unknown='ignore')

ohe.fit(df_train[columns_cat])

df_train[ohe.get_feature_names()] = ohe.transform(df_train[columns_cat]).toarray()
df_test[ohe.get_feature_names()] = ohe.transform(df_test[columns_cat]).toarray()

print('**************************************************************************************************')
print('Checking the column size of Training and Testing data set ')
print('The new columns created by One hot encoding: ', len(ohe.get_feature_names()))
print('Training data set columns: ', df_train.shape)
print('Testing data set columns: ', df_test.shape)
print('**************************************************************************************************')
```

# Feature Selectionl

```python
column_list = columns_num + list(ohe.get_feature_names()) + ['SalePrice']

corr_values = df_train[column_list].corr()['SalePrice']
corr_values

selected_col = list((corr_values[(corr_values > 0.1) | (corr_values <-0.1)]).index)
selected_col.remove('SalePrice')
```

```python
selected_col
```

# Model Training

    1. Model will be trained by below model training techniques

```python
X = df_train[selected_col]
y = df_train['SalePrice']

modelLinear = LinearRegression()
modelLinear.fit(df_train[selected_col],df_train['SalePrice'])

yp = modelLinear.predict(df_test[selected_col])
df_test['SalePrice'] = yp
df_test[['Id','SalePrice']].to_csv('subLR1.csv',index=False)
modelLinear.score(df_train[selected_col],df_train['SalePrice'])
```

```python
modelKNN = KNeighborsRegressor(n_neighbors=3)
modelKNN.fit(df_train[columns_num + list(ohe.get_feature_names())],df_train['SalePrice'])
print(modelKNN.score(df_train[columns_num + list(ohe.get_feature_names())],df_train['SalePrice']))
yp = modelKNN.predict(df_test[columns_num + list(ohe.get_feature_names())])
df_test['SalePrice'] = yp
df_test[['Id','SalePrice']].to_csv('subKNN3.csv',index=False)
```

```python
params = {'n_estimators':[20,30,40,50],'max_depth':[2,3,4,5,6],'min_samples_leaf':[3,4,5,6]}
gridCV = GridSearchCV(RandomForestRegressor(random_state=51),
                     param_grid=params,cv=5,verbose=0,scoring='accuracy')

gridCV.fit(df_train[selected_col],df_train['SalePrice'])
```

```python
gridCV.best_params_

modelRF = RandomForestRegressor(max_depth=12, min_samples_leaf=2,random_state=95)

modelRF.fit(df_train[selected_col],df_train['SalePrice'])
print(modelRF.score(df_train[selected_col],df_train['SalePrice']))

yp = modelRF.predict(df_test[selected_col])

df_test['SalePrice'] = yp
df_test[['Id','SalePrice']].to_csv('modelRF3.csv',index=False)
```

# XGBoost Regressor

```python
modelXG = XGBRegressor()
modelXG.fit(df_train[selected_col],df_train['SalePrice'])

yp = modelXG.predict(df_test[selected_col])

df_test['SalePrice'] = yp
df_test[['Id','SalePrice']].to_csv('modelXG2.csv',index=False)

modelXG.score(df_train[selected_col],df_train['SalePrice'])
```
